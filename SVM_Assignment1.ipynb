Q4. What is the role of support vectors in SVM Explain with example 

```python
# Support vectors play a critical role in Support Vector Machines (SVMs). They are the data points that are closest to the decision boundary (hyperplane) and have a
# direct influence on defining the decision boundary. These support vectors are the most informative data points in the dataset because they determine the position 
# and orientation of the decision boundary.

# Example:
# Let's consider a binary classification problem where you want to distinguish between two classes, A and B, using a linear SVM.

# Data Points: You have a dataset with several data points from both classes, scattered in two-dimensional space. Each data point is represented as (x, y), where x and y
# are the feature values.

# Linear SVM: You decide to use a linear SVM to separate the two classes. In this case, the decision boundary is a straight line.

# Margin: The SVM aims to maximize the margin, which is the distance between the decision boundary and the nearest data points from both classes. The support vectors are 
# the data points that are exactly on the margin or within a margin width of the decision boundary.

# Influence on Decision Boundary: Support vectors have a significant influence on the position and orientation of the decision boundary. In fact, the decision boundary 
# is entirely determined by the support vectors.

# Margin Width: The margin width is twice the distance from the decision boundary to the nearest support vector. Maximizing the margin width ensures a robust and 
# well-generalized model.

# Loss Function: When training an SVM, the objective is to minimize the classification error while maximizing the margin. The margin violations, which are data points 
# that fall inside the margin or on the wrong side of the decision boundary, contribute to the SVM's loss function.
```

Q5. Illustrate with examples and graphs of Hyperplane, Marginal plane, Soft margin and Hard margin in 
SVM?

```python
# 1. Hyperplane:
# The hyperplane is the decision boundary that separates data points of different classes in an SVM. In a two-dimensional space, the hyperplane is a straight line.
#  Here's a visual example:

import matplotlib.pyplot as plt
import numpy as np

np.random.seed(0)
data1 = np.random.randn(50, 2) + np.array([2, 2])
data2 = np.random.randn(50, 2) + np.array([-2, -2])
plt.scatter(data1[:,0], data1[:,1], color='blue', marker='o')
plt.scatter(data2[:,0], data2[:,1], color='red', marker='x')
x = np.linspace(-4, 4, 100)
y = x  
plt.plot(x, y, color='black', linestyle='-')
plt.title('Hyperplane')
plt.xlabel('X1')
plt.ylabel('X2')
plt.show()
```

```python
# 3. Hard Margin:
# A hard margin SVM aims to find a hyperplane that perfectly separates the data with a maximum margin and without any misclassifications.
# Here's a visual example:

# Sample data
np.random.seed(0)
data1 = np.random.randn(50, 2) + np.array([2, 2])
data2 = np.random.randn(50, 2) + np.array([-2, -2])

# Plot the data
plt.scatter(data1[:,0], data1[:,1], color='blue', marker='o')
plt.scatter(data2[:,0], data2[:,1], color='red', marker='x')

# Plot the hard margin hyperplane
x = np.linspace(-4, 4, 100)
y = x  # Assuming a linear hyperplane (slope=1)
plt.plot(x, y, color='black', linestyle='-')

plt.title('Hard Margin SVM')
plt.xlabel('X1')
plt.ylabel('X2')
plt.show()
```

```python
# 4. Soft Margin:
# A soft margin SVM allows for some misclassifications to find a hyperplane with a larger margin. It introduces a penalty for misclassified data points. 
# Here's a visual example:

# Sample data with some overlap
np.random.seed(0)
data1 = np.random.randn(50, 2) + np.array([1, 1])
data2 = np.random.randn(50, 2) + np.array([-1, -1])

# Plot the data
plt.scatter(data1[:,0], data1[:,1], color='blue', marker='o')
plt.scatter(data2[:,0], data2[:,1], color='red', marker='x')

# Plot the soft margin hyperplane
x = np.linspace(-4, 4, 100)
y = x  # Assuming a linear hyperplane (slope=1)
plt.plot(x, y, color='black', linestyle='-')

plt.title('Soft Margin SVM')
plt.xlabel('X1')
plt.ylabel('X2')
plt.show()
```

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline

from sklearn.datasets import load_iris
```

```python
data = load_iris()
```

```python
data.keys()
```

```python
x = pd.DataFrame(data.data,columns=data.feature_names)
y = data.target
```

```python
x.head()
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
from sklearn.svm import SVC
```

```python
svc=SVC()
```

```python
from sklearn.model_selection import GridSearchCV
```

```python
parameter = {
    'C': [0.1, 1, 10],
    'kernel': ['linear', 'rbf', 'poly','sigmoid'],
    'gamma': ['scale', 'auto']
}
```

```python
grid = GridSearchCV(estimator=svc,param_grid=parameter,cv=5,)
```

```python
grid.fit(x_train,y_train)
```

```python
Best_parameter = grid.best_params_
print(Best_parameter)
```

```python
Best_estimetor = grid.best_estimator_
print(Best_estimetor)
```

```python
y_pred = grid.predict(x_test)
y_pred
```

```python
from sklearn.metrics import accuracy_score,confusion_matrix,classification_report
```

```python
print(confusion_matrix(y_pred,y_test))
print(accuracy_score(y_pred,y_test))
print(classification_report(y_pred,y_test))
```
