```python
import bokeh.io
import bokeh.plotting
bokeh.io.output_notebook()
from bokeh.plotting import figure, show
```

```python
# Q1. How can you create a Bokeh plot using Python code?
```

```python
x = [1, 2, 3, 4, 5]
y = [6, 7, 2, 4, 5]
p = figure(title='Bokeh Plot')
p.xaxis.axis_label = 'x'
p.yaxis.axis_label = 'y'
p.circle(x,y, color='blue')
show(p)
```

```python
# Q2. What are glyphs in Bokeh, and how can you add them to a Bokeh plot? Explain with an example.
```

```python
## Glyphs :- In Bokeh, glyphs are visual markers used to represent data points on a plot.
#            Glyphs can be simple shapes such as circles, squares, triangles, or more complex shapes like lines or patches.
#            They allow you to visualize and communicate your data effectively.

from bokeh.plotting import figure, show
x = [1, 2, 3, 4, 5]
y = [6, 7, 2, 4, 5]
p = figure(title='Bokeh Plot', x_axis_label='X', y_axis_label='Y')
p.circle(x, y, size=10, color='blue', alpha=0.5)
p.line(x, y, line_width=2, color='red')
show(p)
```
