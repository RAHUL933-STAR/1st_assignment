Que 1: Name any five plots that we can plot using the Seaborn library. Also, state the uses of each plot.

```python
## Seaborn :- Seaborn is a powerful data visualization library in Python that provides various types of plots to explore and analyze data.

# types of plot:-

# 1) Scatter Plot: A scatter plot displays the relationship between two continuous variables. 
#                  It shows the distribution of individual data points and the correlation between variables.
#  Use cases:- Exploring the correlation between two variables, identifying patterns or clusters in data, and detecting outliers.

# 2) Line Plot:- A line plot shows the relationship between two continuous variables, where one variable represents time or a sequential order.
#                It is often used to visualize trends and patterns over time.
# Use cases: Visualizing time series data, tracking changes in a variable over time, and comparing multiple time series.

# 3) Bar Plot:-  A bar plot represents categorical data with rectangular bars, where the length of each bar corresponds to the value of the category. 
#                It is commonly used to compare and display the distribution of categorical variables.
# Use cases: Comparing the frequency or count of categories, displaying proportions or percentages, and comparing data across different categories.

# 4) Histogram: A histogram displays the distribution of a continuous variable by dividing it into bins and showing the frequency or count of values falling into each bin. 
#               It provides insights into the shape, central tendency, and spread of the data.
# Use cases: Analyzing the distribution of data, identifying skewness or outliers, and exploring the range and frequency of values.

# 5) Heatmap: A heatmap represents data using a grid of colors, where each cell's color indicates the value of a variable. 
#             It is particularly useful for visualizing relationships and patterns in large datasets.
# Use cases: Showing correlation matrices, visualizing intensity or density, identifying clusters or patterns, and analyzing relationships between variables.
```

Que 2: Load the "fmri" dataset using the load_dataset function of seaborn. Plot a line plot using x = 
"timepoint" and y = "signal" for different events and regions. 

Note:  timepoint, signal, event, and region are columns in the fmri dataset.

```python
import seaborn as sns
fmris = sns.load_dataset('fmri')
sns.lineplot(data = fmris, x = 'timepoint', y = 'signal', hue="event", style="region")
```

Que 3: Load the "titanic" dataset using the load_dataset function of seaborn. Plot two box plots using x = 
'pclass', y = 'age' and y = 'fare'.

Note:  pclass, age, and fare are columns in the titanic dataset.

```python
import seaborn as sns
import matplotlib.pyplot as plt
Titanic = sns.load_dataset('titanic')
plt.figure(figsize=(10,6))
plt.subplot(1,2,1)
sns.boxplot(data = Titanic, x = 'pclass', y='age')
plt.title("Box Plot of Age by Passenger Class")
plt.xlabel("Pclass")
plt.ylabel("Age")


plt.subplot(1,2,2)
sns.boxplot(data = Titanic, x = 'pclass', y='fare')
plt.title("Box Plot of Fare by Passenger Class")
plt.xlabel("Pclass")
plt.ylabel("Fare")
plt.show()
```

Que 4: Use the "diamonds" dataset from seaborn to plot a histogram for the 'price' column. Use the hue 
parameter for the 'cut' column of the diamonds dataset. 

```python
import seaborn as sns
import matplotlib.pyplot as plt
plt.figure(figsize=(20,10))
colors = ["#FF0000", "#00FF00", "#0000FF", "#FFFF00", "#FF00FF"]
Diamonds = sns.load_dataset('diamonds')
sns.histplot(data = Diamonds, x = 'price', hue='cut', palette=colors)
```

Que 5: Use the "iris" dataset from seaborn to plot a pair plot. Use the hue parameter for the "species" column 
of the iris dataset.

```python
import seaborn as sns
iris = sns.load_dataset('iris')
sns.pairplot(iris, hue='species')
```

Que 6: Use the "flights" dataset from seaborn to plot a heatmap

```python
import seaborn as sns
import matplotlib.pyplot as plt
Flights = sns.load_dataset('flights')
flights_matrix = flights.pivot("month", "year", "passengers")
sns.heatmap(flights_matrix, annot=True, fmt="d", cmap="YlGnBu")
plt.xlabel("Year")
plt.ylabel("Month")
plt.title("Passenger Count by Year and Month")
plt.show()
```
