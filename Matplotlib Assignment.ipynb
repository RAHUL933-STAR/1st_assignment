Q1: What is Matplotlib? Why is it used? Name five plots that can be plotted using the Pyplot module of 
Matplotlib.

```python
## Matplotlib is a popular Python library used for creating visualizations and plots.
# It provides a wide range of functionalities for generating various types of plots, graphs, charts, and figures.
# Matplotlib is highly customizable, allowing users to create visually appealing and informative visualizations.

## Name of five plots :-

# 1) Line Plot: A line plot is used to display data points connected by straight lines. It is useful for visualizing trends, patterns, or changes over time.

# 2) Bar Plot: A bar plot, also known as a bar chart, is used to compare categorical data by displaying rectangular bars of different heights. 
#              It is commonly used to show comparisons between different groups or categories.

# 3) Scatter Plot: A scatter plot is used to display the relationship between two variables. 
#                 It represents data points as individual dots on a two-dimensional plane, where each dot corresponds to a data point with its x and y values.

# 4) Histogram: A histogram is used to visualize the distribution of a single numerical variable. 
#               It divides the data into bins and displays the frequency or count of data points falling into each bin using rectangular bars.

# 5) Pie Chart: A pie chart is used to represent categorical data as a circular chart. 
#               The size of each "slice" of the pie corresponds to the proportion of each category in the whole data set.
#               It is commonly used to show the composition of a whole in terms of percentages.
```

Q2: What is a scatter plot? Use the following code to generate data for x and y. Using this generated data 
plot a scatter plot.

import numpy as np

np.random.seed(3)

x = 3 + np.random.normal(0, 2, 50)

y = 3 + np.random.normal(0, 2, len(x))

Note: Also add title, xlabel, and ylabel to the plot.

```python
import matplotlib.pyplot as plt
import numpy as np

np.random.seed(3)

x = 3 + np.random.normal(0, 2, 50)

y = 3 + np.random.normal(0, 2, len(x))
plt.scatter(x,y)
plt.xlabel('x axis')
plt.ylabel('y axis')
plt.title('Random Data')
```

Q3: Why is the subplot() function used? Draw four line plots using the subplot() function.

Use the following data:

import numpy as np

For line 1: x = np.array([0, 1, 2, 3, 4, 5]) and  y = np.array([0, 100, 200, 300, 400, 500])

For line 2: x = np.array([0, 1, 2, 3, 4, 5]) and y = np.array([50, 20, 40, 20, 60, 70])

For line 3: x = np.array([0, 1, 2, 3, 4, 5]) and y = np.array([10, 20, 30, 40, 50, 60])

For line 4: x = np.array([0, 1, 2, 3, 4, 5]) and y = np.array([200, 350, 250, 550, 450, 150])

```python
## subplot() function :- The subplot() function in Matplotlib is used to create multiple plots within a single figure. 
#                It allows you to arrange and display multiple plots in a grid-like structure, making it easier to compare and analyze different sets of data.

import numpy as np
import matplotlib.pyplot as plt

x = np.array([0, 1, 2, 3, 4, 5])
y = np.array([0, 100, 200, 300, 400, 500])
plt.subplot(2, 2, 1)
plt.scatter(x, y)
plt.title('Line 1')


x = np.array([0, 1, 2, 3, 4, 5])
y = np.array([50, 20, 40, 20, 60, 70])
plt.subplot(2, 2, 2)
plt.scatter(x, y)
plt.title('Line 2')


x = np.array([0, 1, 2, 3, 4, 5])
y = np.array([10, 20, 30, 40, 50, 60])
plt.subplot(2, 2, 3)
plt.scatter(x, y)
plt.title('Line 3')


x = np.array([0, 1, 2, 3, 4, 5])
y = np.array([200, 350, 250, 550, 450, 150])
plt.subplot(2, 2, 4)
plt.scatter(x, y)
plt.title('Line 4')

plt.show()
```

Q4: What is a bar plot? Why is it used? Using the following data plot a bar plot and a horizontal bar plot. 

import numpy as np

company = np.array(["Apple", "Microsoft", "Google", "AMD"])

profit = np.array([3000, 8000, 1000, 10000])

```python
import matplotlib.pyplot as plt
import numpy as np

company = np.array(["Apple", "Microsoft", "Google", "AMD"])

profit = np.array([3000, 8000, 1000, 10000])

plt.bar(company, profit)
plt.title('Bar plot')
plt.xlabel('company')
plt.ylabel('profit')
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

company = np.array(["Apple", "Microsoft", "Google", "AMD"])

profit = np.array([3000, 8000, 1000, 10000])

plt.barh(company, profit)
plt.title('Bar plot')
plt.xlabel('profit')
plt.ylabel('company')
plt.show()
```

Q5: What is a box plot? Why is it used? Using the following data plot a box plot.

box1 = np.random.normal(100, 10, 200)

box2 = np.random.normal(90, 20, 200)

```python
## A box plot, also known as a box-and-whisker plot, is a graphical representation of the distribution of a dataset.
#  It displays key statistical measures such as the median, quartiles, and potential outliers.

## usage :- Visualize the distribution and spread of data across different categories or groups and Identify skewness, outliers, and potential anomalies 
#           in the dataset.

import matplotlib.pyplot as plt
import numpy as np

box1 = np.random.normal(100, 10, 200)
box2 = np.random.normal(90, 20, 200)
data = [box1, box2]
ax = plt.subplot()
ax.boxplot(data)
ax.set_xticklabels(["Box1", "Box2"])
ax.set_ylabel("Value")
ax.set_title("Box plot")
plt.show()
```
